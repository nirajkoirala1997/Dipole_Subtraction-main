```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Function to read data from a .dat file
def read_data(filename):
    data = np.loadtxt(filename)
    return data
filename1 = 'add.lhc.nlo2.dat'
filename2 = 'add.lhc.nlo3.dat'
```

```python
dir_path = '/home/niraj/OneDrive/WorkSpaceIITG/Generalfiles/integrators/DY1/Ref/Diphoton-Isolation_Deltac/summary/run_01'

# Construct the full file path
file_path1 = os.path.join(dir_path, filename1)
file_path2 = os.path.join(dir_path, filename2)

# Read the data from the files
data1 = read_data(file_path1)
data2 = read_data(file_path2)

#print(data1)
#print(data1)

# Extract the relevant columns
x1, y1 = data1[:, 4], data1[:, 8]
x2, y2 = data2[:, 4], data2[:, 8]

# Plotting the data
plt.figure(figsize=(8, 4))
plt.plot(x1, y1, 'o-', label='nlo2')
plt.plot(x2, y2, 's-', label='nlo3')
plt.xlabel('x (Delta_c)')
plt.ylabel('y (some unit)')
plt.title('Delta_s fixed, ')
plt.grid(True)
plt.legend()
```
